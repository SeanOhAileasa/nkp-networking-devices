###### 2. Firewalls
***

In this presentation, we'll introduce the basic functionality of a firewall, which is software or hardware used to control ingress and egress of network traffic and that refers to the direction. Traffic coming in and/or traffic coming out. 

And you can control both of them with a firewall. And it is quite simply designed to prevent unauthorized access to a private network. Of course on the Internet, there could be intruders that are trying to gain access to proprietary information on the inside. The firewall in essence is designed to protect that information.

Now there are two main types, a software based firewall that runs on a host PC or server. And as such, it's called a host-based firewall.

###### Host-based Firewalls

And it really is just configurable software on that system. And in fact, in Windows environments, all recent Windows clients and servers included a firewall by default ever since Windows XP. And you can just configure it through the Control Panel.

###### Network-based Firewalls
***

And then there's the network-based firewall. This is typically a dedicated device, although they might run on a server, and we'll actually come back to that in a moment. 

But it typically segregates a LAN from devices on an untrusted network. And of course the untrusted network is the Internet. 

So the primary difference is that the network based firewall is designed really to protect the entire network.

###### 3. Routers
***

Now, in this presentation, we'll take a look at the basic characteristics of a router.

And recall from our OSI model that the third layer is the network layer and this is where we find routers. [Router.] Hence, they're often referred to as layer 3 devices. 

< [Table of Contents](#top) | [References](#references) >
<a id="___"></a>
***
###### OSI - Network Layer
***

Now, at the network layer, the devices are not concerned with connecting any specific hosts to each other, just the networks. We also find the logical address at the third layer, in other words, the IP address, in most cases. So routers forward packets off to other networks based on the IP address of the networks that are being connected. But again, it does not concern itself with the host portion of the address. It only considers the network portion of the address, so that it can find the router for that network. And as soon as it forwards the packet off to that next router, itâ€™s job is finished. So again, it's not concerned with finding a specific host on that network, just the way to get in to that network. So it's often referred to as a gateway device because it sits on the border or the boundary of your network. And it's our last point of contact as packets move out of the network. And it's the first point of contact for packets that are coming into our network. 

So it always connects at least two networks together. And that can be a LAN to a WAN, a LAN to another LAN within your own organization, or of course, very commonly, your LAN to your Internet service provider. 

Now, routers can contain processors, memory, and various input or output interfaces because they need to make decisions, if you will. There needs to be some programming in a router. They are effectively smarter than devices such as hubs, they don't make any decisions. But routers need to be programmed, so there has to be a processor, there needs to be memory. And the interfaces are how we configure the device. So it doesn't have its own keyboard or display. You connect to one of those I/O interfaces using your own computer, and you configure it remotely. 

Typically, most higher level routers are just rack-mounted devices that, again, just don't have any kind of configuration capabilities inherently within the device itself. Maybe you might be able to reset it back to its factory default settings, that's about it. Almost every other configuration is done using a remote connection through one of those interfaces.

###### 4. Switches and Hubs
***

Now taking a closer look at some of the characteristics of the lower layer devices, if we go all the way to the physical layer, or Layer 1, [Hub.] we find the hub. 

< [Table of Contents](#top) | [References](#references) >
<a id="___"></a>
***
###### Hub
***

Now this is used to connect hosts together or segments of a LAN. 

And it contains multiple ports, and packets arriving at one port are copied to all other ports. So this does allow all segments to see those packets or all hosts to see those packets. But, of course, the problem is, at the physical layer there is no addressing whatsoever. So quite simply stated, hubs aren't particularly smart. They are not aware of any kind of address. Despite the fact that the MAC address is often referred to as a physical address, it resides at Layer 2. So because a hub is unaware of any kind of address, anything that comes in one port has to be copied out every other port. Because the hub has no way to distinguish which system is connected to any given ports. So it has to copy that information to every port. So quite simply, hubs are noisy. They did the job, but they caused a lot of unnecessary traffic. Now their attractive feature, going back several years, was quite simply the fact that they were less expensive than switches. But in today's networking environments, it would be very rare to find a hub.

< [Table of Contents](#top) | [References](#references) >
<a id="___"></a>
***
###### Switch
***

So switches, quite simply, operate at the next layer up. They operate at the data link layer, and as such, they are aware of MAC addresses. 

Recall, the data link layer is subdivided into the MAC sublayer and logical link control.

So now we start to see an address implemented. And because of that, we can increase the overall bandwidth of the environment simply by not flooding the network with a bunch of unnecessary traffic. 

So switches are aware that computer one is connected to physical port one and computer two is connected to physical port two and so on. 

As packets arrive from that system, it picks up on the MAC address that was used to send it and it records it, and it now knows that this system resides on that port. Now, you can end up with certain scenarios where you have multiple switches connected to each other. Whereby any given system is, at least temporarily, unable to find any other system, perhaps because those tables have been cleared or changed, or anything along those lines. So it is possible that you can get these looping scenarios. But that can be addressed and/or corrected through the use of VLANs and what's known as the Spanning Tree Protocol. And we'll talk about some of those configurations in greater detail in some upcoming presentations. But this, quite simply, prevents that looping from happening, or at least if it does happen, it detects it and is able to stop it from continually looping. 

< [Table of Contents](#top) | [References](#references) >
<a id="___"></a>
***
###### Switch Configuration
***

Now not only is a switch capable of filtering traffic based on MAC addresses, it also supports configuration which, again, is something that hubs do not support. 

So common Ethernet switch port parameters include the interface speed, the duplex, and the auto mdix, which is media dependent interface crossover. Now we'll come back to that in the moment.

The speed is, quite simply, how fast can this physical port communicate with whatever is on the other end. 

Now most switches these days, I might even go so far as to say all of them, would have an auto-sensing capability so that it can negotiate. But if you know that the device on the other end can only communicate, let's say, at a 100 megabits per second, maybe it's a little bit older, then you can statically set that and you reduce a little bit of processing. Because the auto-sensing does have to do a little bit of negotiation, if you will. 

The same goes for the duplex, that's whether or not it can send and receive at the same time. So full duplex is the ability to do just that, but half duplex is send and receive separately. So you can configure that as well. And, again, there would be the auto-sensing capability. But, as mentioned, the less you have to auto-sense, the less processing you have to do. So static configuration typically is better but, of course, if something changes, you have to go back in and reconfigure. 

< [Table of Contents](#top) | [References](#references) >
<a id="___"></a>
***
###### Switch Configuration - Auto MDIX (Media Dependent Interface Crossover)
***

Now the auto mdix, again, is that media dependent interface crossover, that deals with the wire configuration in terms of transmitting and receiving. For example, a regular Ethernet port uses the pins 1 and 2 to transmit and 3 and 6 to receive. So that is a function, then, of what type of physical cable you place between the two endpoints. So if it's a regular computer on one side, then the switch on the other, that's the standard straight-through cable and the pins are placed accordingly to communicate. 

But if it's switch to switch, that is typically where you see the crossover cable, or the rollover cable depending on the circumstances, but they have different wire configurations. But, again, a lot of newer devices have this auto mdix capability, which means that it can digitally rework the connection, if you will, so that you can actually use a straight-through cable in a scenario where it expects to see a crossover or a rollover. So it quite literally reconfigures the wires digitally, if you will, before the communications are sent to the processor. 

So it doesn't really matter how the physical wire is lined up, in that case. But, again, it is something that you can set statically or you can leave it set to this auto-sensing capability, so that it can determine it for itself. 

Ultimately, the auto-sensing capability reduces the changes that need to be made, but it involves more processing. So you might be able to improve the overall performance of the switch by statically configuring these values, but that introduces a little more overhead if and/or when things change.

Well, there are a few scenarios where they do differ, and weâ€™ll come to those. But before that, I just want to stress the keyword of network segment here. When you are talking about using a bridge, you aren't connecting networks together like a router. Recall, the router works at the network layer. So it performs the same type of functionality in terms of trying to keep the traffic separate on either side, but it still does so based on MAC addresses. ###### 6. Modems

Our next interconnectivity device is the modem.

[Modem.] Although, in certain very rural areas, you may still see them used to connect to your Internet service provider, simply because they don't have the infrastructure in places so far out. But very rare by today's standards. It is short for modulator-demodulator and quite simply enables a computer to transmit data using a telephone line. 

And of course, that was the key component back in the early days of Internet connectivity. We didn't have high-speed networks all over the globe. But we did have the phone line. It was already there, so the infrastructure was already in place. So we could simply take advantage of that to communicate. The standard interface was called an RS-232. 

Now, clearly, the speeds that we need today were not supported by modems. The phone lines could only carry very small amounts of data. There was very slow bandwidth. And that simply means that the signal that is sent out uses digits, ones and zeros, to create a waveform analogous to the actual electrical current. That's where the term analog came from. It is analogous to the actual signal. So again, we are feeding ones and zeros into the modem. The modem has to modulate/demodulate those ones and zeros into electrical signals. So that's how the communication was facilitated, but again, it was, of course, very slow. 

But it did have the advantage of global access pretty much right from the get go, because the phone lines were available just about globally.

###### 7. Wireless Routers and Access Points
***

Now in this presentation, we'll take a closer look at wireless routers and access points, which are special purpose communication devices for wireless LANs or WLANs. The access points are used to provide a hotspot for wireless clients in the vicinity. ###### 9. Wireless Range Extenders
***

Now, this presentation will take a look at the wireless range extender. [Wireless Range Extender.] Which, as its name indicates, is used to extend the radio frequency range of wireless transmissions.

Now, it functions down at layer 1, the physical layer of the OSI model. And recall that in essence, there's not a whole lot of intelligence happening down at layer 1. There is really nothing more happening than hearing the signal on one side, boosting it, or what's also known as repeating it, and sending it along its way. So it allows you to extend the distance that can be covered of any given Ethernet network segment but, in this case, it's just done wirelessly. 

So again, there's no intelligence happening here. It simply hears something coming in, boosts it, and sends it along its way. And then, of course, does the exact the same thing in reverse with respect to the devices that are communicating with it on the other side.

###### Distance
***

So in terms of the distance, it does depend on several factors, the transmission power of the access point itself and the gain of the antenna that is used. And, of course, the clients also are transmitters and receivers. So while this can certainly boost the range of the signal in terms of you being able to get farther away from the original router, but you still have to remain within a certain distance of the extender as well. Because, of course, your device has to transmit signals back to the extender. So your device does not get any more powerful because of the range extender. ###### Connectivity
*** 

In terms of connectivity, it allows computers and other devices to connect to a wireless LAN generally outside of the wireless access point's normal range. So again, that is referring to the router itself. If that has a distance of X, then the extender can maybe go to 2X or 3X, depending on the power. 

But that, again, only allows you to get a little farther away from the original router itself. You still have to be within range of the extender.
